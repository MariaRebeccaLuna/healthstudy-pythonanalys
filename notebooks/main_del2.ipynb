## Utökad analys och visualisering:

Jag valde att visa/räkna sjukdomsförekomst per kön som en ny visualisering.

```python
import matplotlib.pyplot as plt

df = pd.read_csv("../data/health_study_dataset.csv")

disease_by_sex = df.groupby("sex")["disease"].mean()

plt.figure()

disease_by_sex.plot(kind="bar")

plt.ylabel("Andel med sjukdom")
plt.title("Sjukdomsförekomst per kön")

plt.show()
```
